```python
#Python Lists Vs Numpy Arrays
import time
```

```python
py_list = list(range(1000000))
start = time.time()
multiplied = [x*2 for x in py_list]
t = time.time() - start
print(t)

arr = np.arange(1000000)
begin = time.time()
marr = arr *2
end = time.time() - begin
print(end)#20-50x faster
```

```python
#creating arrays from lists
a = np.array([1 , 2 , 3])#1d arrays from a list manual creation
b = np.array([1 , 4 , 5 ,6 , 6])
print(a)
print(b)
```

```python
#creating arrays from builtin method
a = np.zeros((2 , 3))
b = np.ones((2 , 3))
c = np.eye(3)
d = np.full((2 , 2) , 7)
e = np.arange(1 , 1000 , 2)
f = np.random.rand(2 , 2)
g = np.linspace(0 ,1 , 5)
print(a , b , c , d , e ,f ,g)
```

```python
#Properties Of Arrays
a = np.array([1 , 2 , 3] , dtype='float64')#We Can Force
print(a.shape)
print(a.size)
print(a.ndim)
print(a.dtype)
```

```python
#Reshaping Arrays
a=np.arange(2 , 12)
a.reshape(2,5)
a.reshape(-1 , 2)
x = a.flatten()#copy Is Created No Change to original array
print(x)
x[0] = 999
print(a)
y = a.ravel()#replaces original
y[0] = 999
print(y)
print(a)
```

```python
a= np.array([[1, 2 , 3] , [4 , 5 , 6]])
#normal indexing
print(a[0])
print(a[1 , 2])
#Boolean Indexing
print(a[a>3])#Under A Rule
#Fancy Indexing
b = np.array([1 , 2 , 4])
print(b[[0 , 1 , 2]])
```
